```python
import pandas as pd
```

```python
data1 = pd.read_csv("file1.csv")
data2 = pd.read_csv("file2.csv")
data3 = pd.read_csv("file3.csv")
print(data1.columns)
print(data2.columns)
print(data3.columns)
```

```python
def standardize(df):
    columns_lower = []

    for column in df.columns:
        columns_lower.append(column.lower().replace(" ", "_"))
    df.columns = columns_lower 

    return df 
```

```python
standardize(data1)
```

```python
standardize(data2)
```

```python
standardize(data3)
```

```python
data1.drop("customer", axis=1, inplace=True)
data2.drop("customer", axis=1, inplace=True)
data3.drop("customer", axis=1, inplace=True)
```

```python
data2 = data2.reindex(columns=["st", "gender", "education", 'customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount'])
```

```python
data3 = data3.reindex(columns=["st", "gender", "education", 'customer_lifetime_value', 'income', 'monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount'])
```

```python
data_result = pd.concat([data1, data2, data3])
```

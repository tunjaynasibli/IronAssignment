```python
import config
```

```python
import spotipy
import pandas as pd

from spotipy.oauth2 import SpotifyClientCredentials

```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=config.client_id, client_secret=config.client_secret))
```

```python
song = sp.search(q="Bad Guy", limit=50,market="GB") 
song
```

```python
sp.audio_features(song["tracks"]["items"][0]["id"] )
```

```python
#my_dict = sp.audio_features(song["tracks"]["items"][0]["uri"])[0] # you can provide a list of uri's

my_music_data=[]
for index in range(0,len(song["tracks"]["items"])):
    my_music_data.append(sp.audio_features(song["tracks"]["items"][index]["uri"])[0])
pd.DataFrame(my_music_data)  
```

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
song2 = get_playlist_tracks("spotify", "4rnleEAOdmFAbRcNCgZMpY")
```
